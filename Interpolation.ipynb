```python
from matplotlib import pyplot as plt
import numpy as np
import matplotlib
import math
```

```python
def f(x):
    return np.power(x, 3) - np.power(math.e, x) + 1
```

```python
a = -5
b = 5
```

```python
def l(x, node, k, n):
    ans = 1
    for i in range(n+1):
        if (i!=k):
            ans *= (x - node[i] )/ (node[k] - node[i])
    return ans

def L_n(x, node, n):
    ans = 0
    for i in range(n+1):
        ans += l(x, node, i, n) * f(node[i])
    return ans

def cheb_node(a, b, n):
    nodes = []
    for i in range(n+1):
        x = 1/2*((b-a)*math.cos((2*i+1)/(2*(n+1))*math.pi)+(b+a))
        nodes.append(x)
    return nodes
    
plt.figure(figsize=(12, 6))
x = np.linspace(a,b,10000)
y = [f(i) for i in x]
plt.plot(x,y, label="f(x)")

node = np.linspace(a,b,3)
n = 2
y = [L_n(i, node, n) for i in x]
plt.plot(x,y, label= "L,n=2")    

node = np.linspace(a,b,4)
n = 3
y = [L_n(i, node, n) for i in x]
plt.plot(x,y, label= "L,n=3")    

node = np.linspace(a,b,5)
n = 4
y = [L_n(i, node, n) for i in x]    
plt.plot(x,y, label= "L,n=4")

plt.legend(loc='lower right')
plt.show()


```
